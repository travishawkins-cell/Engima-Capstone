## Simple pdf loader and output generator

```python
import numpy as np
import string

# Dictionary to map digits to their spelled-out forms
digit_map = {
    '0': 'zero', '1': 'one', '2': 'two', '3': 'three', '4': 'four',
    '5': 'five', '6': 'six', '7': 'seven', '8': 'eight', '9': 'nine'
}

# Load the text file
with open('bee movie script.txt', 'r') as file:
    text = file.read()

# Replace all digit characters with their spelled-out forms
processed_text = ''
for char in text:
    if char.isdigit():
        processed_text += digit_map[char]
    else:
        processed_text += char

# Create a list of all characters
char_list = list(processed_text)

# Convert to numpy matrix with 1 column
matrix = np.array(char_list).reshape(-1, 1)

# Display the matrix
print("Matrix shape:", matrix.shape)
print("\nFirst 20 rows of the matrix:")
print(matrix[:20])

# Save the matrix as a text file
np.savetxt('output_matrix.txt', matrix, fmt='%s', delimiter='')

print("\nMatrix saved to 'output_matrix.txt'")

# Create version without punctuation
char_list_no_punct = [char for char in processed_text if char not in string.punctuation]
matrix_no_punct = np.array(char_list_no_punct).reshape(-1, 1)

# Save the no-punctuation version
np.savetxt('output_matrix_no_punctuation.txt', matrix_no_punct, fmt='%s', delimiter='')

print("Matrix without punctuation saved to 'output_matrix_no_punctuation.txt'")

# Create version without spaces
char_list_no_spaces = [char for char in processed_text if char != ' ']
matrix_no_spaces = np.array(char_list_no_spaces).reshape(-1, 1)

# Save the no-spaces version
np.savetxt('output_matrix_no_spaces.txt', matrix_no_spaces, fmt='%s', delimiter='')

print("Matrix without spaces saved to 'output_matrix_no_spaces.txt'")

print(f"\nOriginal matrix rows: {matrix.shape[0]}")
print(f"No-punctuation matrix rows: {matrix_no_punct.shape[0]}")
print(f"No-spaces matrix rows: {matrix_no_spaces.shape[0]}")
```
